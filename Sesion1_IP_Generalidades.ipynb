## **Calculadora de Promedios de Notas.**
Vamos a crear una calculadora para saber cuál es nuestro promedio después de haber acabado semestre.

---

Supongamos que terminaste tu semestre en el cual cursaste 5 materias y obtuviste las siguientes notas:
* 3,  4.3,  2.8,  3.5,  4.7

Queremos saber cuál fue nuestro promedio, para ello debemos sumar todas las notas y dividirlas entre 5.

**For**

Empezamos definiendo un loop. La idea es hacer que una parte de nuestro código se ejecute varias veces para no tener que hacerlo nosotros. Para esto usamos un *for*.

**If**

Definimos ahora un condicional. La idea es ponerle condiciones a ciertas partes de nuestro código para que dependiendo de estas condiciones, dicha parte del código se ejecute o no.


````
notas = [3, 3.3, 2.3, 3.5, 2.7]
total_notas = len(notas)
suma = 0
for i in notas:
  suma = suma + i
promedio = suma/total_notas
print(promedio)
if promedio >= 3:
  print('Aprobé el semestre!')
else:
  print('Creo que tenemos un problema.')
````
